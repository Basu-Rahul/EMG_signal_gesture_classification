```python
from statistics import variance as var
from statistics import mean as mean

def rms(x):
    # x is array of 20 elements
    r,s=0,0
    for i in range(20):
        s+=x[i]**2
    r=np.sqrt(s/2)
    return r

def iemg(x):
    # x is array of 20 elements
    s=0
    for i in range(20):
        s+=x[i]
    #r=np.sqrt(s/2)
    return s
```

```python
def feature_extract(k):    
    st=0
    k=k-1
    r=np.zeros((11,5,20),dtype=np.float32)
    v=np.zeros((11,5,20),dtype=np.float32)
    m=np.zeros((11,5,20),dtype=np.float32)
    e=np.zeros((11,5,20),dtype=np.float32)
    w=np.zeros((11,5,20))
    print(p)
    for i in range(5):
        st=0
        for j in range(p):
            x=list(p2[sample_legend[i+1]+str(2)][st:st+l])
            #print(x, st)
            r[k][i][j]=rms(x)
            #print(r[k][i][j])
            v[k][i][j]=var(x)
            #print(v[k][i][j])
            m[k][i][j]=mean(x)
            #print(m[k][i][j])
            e[k][i][j]=iemg(x)
            #print(e[k][i][j])
            w[k][i][j]=len(x)
            st=st+l
            if (p-j)==1:
                break
    return r, v, m, e, w    
```

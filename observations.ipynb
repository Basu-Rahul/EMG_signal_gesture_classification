```python
import pandas as pd
import matplotlib.pyplot as plt

file=pd.read_csv(r'C:\rahul\rahul\NITK\Semester IV\IET_project\samples\samples2.csv')
```

```python
plt.plot(file['op_cl_op1'])
```

```python
columns_to_plot=['op_cl_op1','op_cl_op3','op_cl_op5']
# Create the plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
for col in columns_to_plot:
  plt.plot(file.index[100:399], file[col][100:399], label=col)  # Plot each column with label

# Add labels and title
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Time domain graph for some subjects for Open Close Open gesture')

# Add legend
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

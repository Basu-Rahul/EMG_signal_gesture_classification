```python
pi=np.pi
def DTFT(N):
    w=np.zeros((N,N),dtype=np.complex_)
    for k in range(N):
        for n in range(N):
            w[k][n]=np.exp(-1j*2*pi/N*n*k)
    return w

def IDTFT(N):
    w=np.zeros((N,N),dtype=np.complex_)
    for k in range(N):
        for n in range(N):
            w[k][n]=np.exp(1j*2*pi/N*n*k)
    return w/N
```
